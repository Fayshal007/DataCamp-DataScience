## Find the mean, median, minimum, maximum, and standard deviation of order_amount

```python
print("mean order_amount:", orders['order_amount'].mean())
print("median order_amount:", orders['order_amount'].median())
print("minimum order_amount:", orders['order_amount'].min())
print("maximum order_amount:",orders['order_amount'].max())
print("standard deviation of order_amount",orders['order_amount'].std())
```

## Find the earliest and latest order_date

```python
print("earliest order_date", orders["order_date"].min())
print("latest order_date", orders["order_date"].max())
```

## Create a function pct30() that returns the 30th percentile of a column, then use .agg() to find the 30th percentile of order_amount.

```python
def pct30(column): 
    return column.quantile(0.3) 
```

```python
orders['order_amount'].agg(pct30)
```

## Use .agg() to calculate the mean and median of both order_amount and delivery_days

```python
orders[['order_amount', 'delivery_days']].agg([np.mean, np.median])
```
